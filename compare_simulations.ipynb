```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# Figure setup
# Set the font to Courier
mpl.rcParams['font.family'] = 'Courier'

# Set the line weight to thin
mpl.rcParams['lines.linewidth'] = 0.75

# Set the color scheme to black and white
mpl.rcParams['text.color'] = 'black'
mpl.rcParams['axes.labelcolor'] = 'black'
mpl.rcParams['xtick.color'] = 'black'
mpl.rcParams['ytick.color'] = 'black'
mpl.rcParams['axes.edgecolor'] = 'black'
```

```python
import pandas as pd
import os

def load_simulation_data(folder_path):
    """Load all pickled DataFrames from the specified folder."""
    data_frames = {}
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.pkl'):
            file_path = os.path.join(folder_path, file_name)
            df = pd.read_pickle(file_path)
            key_name = file_name.replace('.pkl', '')
            data_frames[key_name] = df
    return data_frames

# Example usage
folder_path = 'DEMO'  # Replace with your folder path
data_frames = load_simulation_data(folder_path)
print(data_frames.keys())  # Print the keys to verify the loaded DataFrames
```

```python
def plot_total_sales(data_frames, simulations, brands):
    """Plot total sales for the specified simulations and brands."""
    plt.figure(figsize=(15, 6))
    
    for sim in simulations:
        if sim in data_frames:
            df = data_frames[sim]
            for brand in brands:
                if f'Total_Sales_{brand}' in df.columns:
                    plt.plot(df['Date'], df[f'Total_Sales_{brand}'], label=f'{sim} - {brand}')
    
    plt.title('Total Sales Comparison')
    plt.xlabel('Date')
    plt.ylabel('Total Sales')
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.10), ncol=2, fontsize='small', frameon=False)
    plt.show()
```

```python
plot_total_sales(data_frames, ['DEMO_Nil'],['A','B'])
```
