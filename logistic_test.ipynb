```python
import numpy as np
```

```python
def percent_to_probability(percent, k=11, x0=0.25):
    """
    Computes the probability using a logistic function based on the input percent.

    Parameters:
    percent (float or np.ndarray): The input percent expressed as a decimal (e.g., 0.15 for 15%).
    k (float): The steepness of the logistic curve. Default is 3.
    x0 (float): The midpoint of the logistic curve. Default is 0.1.

    Returns:
    float or np.ndarray: The probability ranging from 0 to 1.
    """
    percent = np.asarray(percent)  # Ensure percent is a NumPy array
    probabilities = np.where(
        (percent == 0.0) | (percent == 1.0),
        0,
        np.where(
            percent <= 1.0,
            1 / (1 + np.exp(-k * (percent - x0))),
            1 / (1 + np.exp(k * (percent - 1.0)))
        )
    )
    return probabilities
```

```python

depths = np.linspace(0.,1.9,100)
```

```python
probs = percent_to_probability(depths)
```

```python
import matplotlib.pyplot as plt

plt.plot(depths, probs, label='Probability Curve')
plt.xlabel('Depths')
plt.ylabel('Probability')
plt.title('Probability vs Depths')
plt.legend()
plt.grid(True)
plt.show()
```
